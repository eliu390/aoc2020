```python
def aoc2_solve(part):
    with open("aoc2_data.txt") as f:
        count = 0
        for line in list(f):
            if valid_pw(part, line):
                count += 1
        return count
    
def valid_pw(part, string):
# boolean validates password string
# passwords resemble: '4-8 g: ggtxgtgbg'
    split1 = string.split('-')
    a = int(split1[0])
    split2 = split1[1].split(' ')
    b = int(split2[0])
    char = split2[1].split(':')[0]
    pw = split2[2]
    
    count = 0
    if part == 1:
        for letter in pw:
            if letter == char:
                count += 1
        return count >= a and count <= b
    if part == 2:
        left = pw[a - 1] == char
        right = pw[b - 1] == char
        return left != right
```

```python
def aoc4_solve(part):
    with open('aoc4_data.txt') as f:
        passports = f.read().split('\n\n')
        count = 0
        for passport in passports:
            if valid_passport(part, passport):
                count += 1
        return count
    
def valid_passport(part, string):
    check = { 'byr': None, 'iyr': None, 'eyr': None, 'hgt': None, 'hcl': None, 'ecl': None, 'pid': None }
    fields = string.replace('\n', ' ').split(' ')
    for field in fields:
        [field, value] = field.split(':')
        check[field] = value
    
    count = 0
    for index, (key, value) in enumerate(check.items()):
        if value == None:
            return False
        if part == 2:
            if valid_field(key, value) == False:
                return False
    return True

def valid_field(key, value):
    if key == 'byr':
        try:
            num = int(value)
            return 1920 <= num <= 2002
        except:
            return False
    if key == 'iyr':
        try:
            num = int(value)
            return 2010 <= num <= 2020
        except:
            return False
    if key == 'eyr':
        try:
            num = int(value)
            return 2020 <= num <= 2030
        except:
            return False
    if key == 'hgt':
        try:
            unit = value[-2:]
            num = value[:-2]
            if unit == 'cm':
                return 150 <= int(num) <= 193
            elif unit == 'in':
                return 59 <= int(num) <= 76
            return False
        except:
            return False
    if key == 'hcl':
        if len(value) != 7:
            return False
        if value[0] != '#':
            return False
        try:
            int(value[1:], 16)
        except:
            return False
        return True
    if key == 'ecl':
        colors = [ 'amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth' ]
        try:
            colors.index(value)
            return True
        except:
            return False
    if key == 'pid':
        if len(value) != 9:
            return False
        try:
            num = int(value)
            return num <= 999999999
        except:
            return False
```

```python
def aoc5_solve(part):
    with open('aoc5_data.txt') as f:
        lines = list(f)
        if part == 1:
            track = []
            for line in lines:
                [ row, col ] = gen_seat_id(line)
                track.append(8 * row + col)
            return max(track)
        elif part == 2:
            track = [ [ 0 for i in range(8) ] for j in range(128) ]
            for line in lines:
                [ row, col ] = gen_seat_id(line)
                track[row][col] = 1
            for row_index, row in enumerate(track):
                for col_index, col in enumerate(row):
                    if col_index == 0 or col_index == 7:
                        continue
                    elif col == 0 and row[col_index - 1] == 1 and row[col_index + 1] == 1:
                        return 8 * row_index + col_index

def gen_seat_id(string):
    min_row = min_col = 0
    max_row = 127
    max_col = 7
    for char in string:
        if char == 'F':
            max_row = round((min_row + max_row + 1) / 2) - 1
        elif char == 'B':
            min_row = round((min_row + max_row + 1) / 2)
        elif char == 'L':
            max_col = round((min_col + max_col + 1) / 2) - 1
        elif char == 'R':
            min_col = round((min_col + max_col + 1) / 2)
    return [ max_row, max_col ]
```

```python
def aoc8_solve1():
    with open('aoc8_data.txt') as f:
        lines = list(f)
        acc = index = 0
        while True:
            # visited = return acc
            if lines[index] == 'x':
                return acc
            
            # parse rule
            line = lines[index].split(' ')
            num = int(line[1][1:])
            if line[1][0] == '+':
                sign = 1
            else:
                sign = -1
                
            # mark line as visited
            lines[index] = 'x'
            
            # execute rule
            if line[0] == 'acc':
                acc += sign * num
                index += 1
            elif line[0] == 'jmp':
                index += sign * num
            else:
                index += 1
```

```python
def aoc11_solve1():
    def adjacent(row, col, lines):
        output = {'.': 0, 'L': 0, '#': 0}
        top = row == 0
        bottom = row == len(lines) - 1
        left = col == 0
        right = col == len(lines[0]) - 1
        if not left:
            output[lines[row][col - 1]] += 1
            if not top:
                output[lines[row - 1][col - 1]] += 1
            if not bottom:
                output[lines[row + 1][col - 1]] += 1
        if not right:
            output[lines[row][col + 1]] += 1
            if not top:
                output[lines[row - 1][col + 1]] += 1
            if not bottom:
                output[lines[row + 1][col + 1]] += 1
        if not top:
            output[lines[row - 1][col]] += 1
        if not bottom:
            output[lines[row + 1][col]] += 1
        return output
        
    def evolve(lines):
        max_row = len(lines)
        max_col = len(lines[0])
        output = [[0] * max_col for i in range(max_row)]
        changed = False
        total = 0
        for row in range(max_row):
            for col in range(max_col):
                if lines[row][col] == '.':
                    output[row][col] = '.'
                elif lines[row][col] == 'L':
                    neighbors = adjacent(row, col, lines)
                    if neighbors['#'] == 0: # success - no occupied neighbors
                        output[row][col] = '#'
                        changed = True
                        total += 1
                    else: # failure - seat remains empty
                        output[row][col] = 'L'
                elif lines[row][col] == '#':
                    neighbors = adjacent(row, col, lines)
                    if neighbors['#'] >= 4: # success - >= 4 occupied neighbors
                        output[row][col] = 'L'
                        changed = True
                    else: # failure - seat remains occupied
                        output[row][col] = '#'
                        total += 1
                else:
                    print('problem:', row, col)
        return total, changed, output
        
    with open('aoc11_data.txt') as f:
        lines = [list(line.strip('\n')) for line in list(f)]
        (total, changed, output) = evolve(lines)
        while changed:
            (total, changed, output) = evolve(output)
        print(total, changed)
        for line in output:
            print(''.join(line))
```
